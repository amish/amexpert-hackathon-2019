```python
import pandas as pd
```

```python
cdf = pd.DataFrame(pd.read_csv('data/campaign_data.csv'))
#campaign_id 	Unique id for a discount campaign
```

```python
print(cdf.shape)
print(cdf.dtypes)
cdf.head()
```

```python
cdf['sdate'] = pd.to_datetime(cdf['start_date'], format='%d/%m/%y', dayfirst=True)
cdf['edate'] = pd.to_datetime(cdf['end_date'], format='%d/%m/%y', dayfirst=True)
cdf.head()
```

```python
cdf['date_diff'] = (cdf['edate'] - cdf['sdate']).dt.days
cdf.head()
```

```python
cdf.to_csv('updated-data/updated_campaign_data.csv', columns={'campaign_id','campaign_type','date_diff'}, index=False)
```
