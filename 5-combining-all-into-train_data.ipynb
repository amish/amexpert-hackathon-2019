```python
train = pd.DataFrame(pd.read_csv('data/train.csv'))
coupm = pd.DataFrame(pd.read_csv('data/coupon_item_mapping.csv'))

campd = pd.DataFrame(pd.read_csv('updated-data/updated_campaign_data.csv'))
custd = pd.DataFrame(pd.read_csv('updated-data/updated_cd.csv'))
custt = pd.DataFrame(pd.read_csv('updated-data/updated_ctd.csv'))
itemd = pd.DataFrame(pd.read_csv('updated-data/updated_item_data.csv'))
```
